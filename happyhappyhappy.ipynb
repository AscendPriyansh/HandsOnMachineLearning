```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
url = "https://ourworldindata.org/grapher/happiness-cantril-ladder.csv?v=1&csvType=full&useColumnShortNames=false"
storage_options = {'User-Agent': 'Mozilla/5.0'}
df = pd.read_csv(url, storage_options=storage_options)
df.columns = ['Country', 'Code', 'Year', 'Life_Satisfaction']
```

```python
real_data = {
    'Country': ['India', 'Nigeria', 'Russia', 'Brazil', 'United States', 'Australia', 'Norway', 'Switzerland', 'Luxembourg', 'Mexico', 'France', 'South Africa'],
    'gdp_per_capita': [9818, 7994, 41705, 19652, 75489, 60304, 63585, 68393, 110261, 17887, 45000, 11466],
    'OBS_VALUE': [4.39, 4.88, 5.95, 6.49, 6.72, 6.97, 7.60, 7.50, 6.90, 6.50, 6.60, 4.70]
}
```

```python
lifeSat = pd.DataFrame(real_data)
```

```python
X = lifeSat[["gdp_per_capita"]].values
Y = lifeSat[["OBS_VALUE"]].values
```

```python
#Visualizing
lifeSat.plot(kind='scatter', grid=True, x="gdp_per_capita", y="OBS_VALUE", color="darkcyan", s=50)

plt.title("GDP per Capita vs. Life Satisfaction", fontsize=14, pad=15)
plt.xlabel("GDP per Capita (USD)", fontsize=11)
plt.ylabel("Life Satisfaction (0 - 10 Scale)", fontsize=11)
```

```python
#Select a linear model
model = LinearRegression()
```

```python
#Train the model
model.fit(X, Y)
```

```python
#Make a Prediction for India
X_new = [[45990]]
print(model.predict(X_new))
```

```python
model = KNeighborsRegressor(n_neighbors=3)
model.fit(X, Y)
```

```python
X_new = [[45990]]
print(model.predict(X_new))
```
